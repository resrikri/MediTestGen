```python


!pip install -q pandas ipywidgets

import pandas as pd
from IPython.display import display, HTML, FileLink
import ipywidgets as widgets

requirements = [
    {"id": "REQ001", "text": "Patient data must be encrypted using AES-256 and comply with GDPR."},
    {"id": "REQ002", "text": "System should allow only authorized healthcare providers to access patient records."},
    {"id": "REQ003", "text": "All diagnostic reports must be stored securely and maintain audit logs."}
]

df = pd.DataFrame(requirements)

def generate_test_case(requirement_text):
    return f"Test Case: Verify that '{requirement_text}' is correctly implemented."

regulations = ["GDPR", "HIPAA", "ISO 27001"]

def add_traceability(df):
    df['test_case'] = df['text'].apply(generate_test_case)
    df['regulation'] = [regulations[i % len(regulations)] for i in range(len(df))]
    return df

df = add_traceability(df)

def on_submit(change):
    global df
    new_req = text_input.value.strip()
    if new_req != "":
        new_id = f"REQ{len(df)+1:03d}"
        new_row = {"id": new_id, "text": new_req}
        df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
        df = add_traceability(df)
        output_area.clear_output()
        with output_area:
            display(HTML("<h3>✅ New requirement added and test case generated:</h3>"))
            display(df.tail(1))
        text_input.value = ""

def download_csv(df):
    filename = "MediTestGen_Dataset.csv"
    df.to_csv(filename, index=False)
    display(FileLink(filename))

text_input = widgets.Text(
    value='',
    placeholder='Enter new healthcare requirement',
    description='Requirement:',
    layout=widgets.Layout(width='70%')
)

submit_button = widgets.Button(description="Generate Test Case", button_style='success')
submit_button.on_click(lambda x: on_submit(None))

download_button = widgets.Button(description="Download CSV", button_style='info')
download_button.on_click(lambda x: download_csv(df))

output_area = widgets.Output()

display(HTML("<h2 style='color: #2F4F4F;'>MediTestGen: AI-Powered Test Case Generator</h2>"))
display(HTML("<p>Enter a new healthcare requirement below and click 'Generate Test Case' to see AI-generated output with compliance traceability.</p>"))
display(widgets.HBox([text_input, submit_button]))
display(output_area)
display(HTML("<h3>Complete Dataset with Traceability:</h3>"))
display(df)
display(download_button)
```
